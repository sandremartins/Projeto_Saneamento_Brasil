```python
# Importe as bibliotecas necessárias para este projeto
import numpy as np
import pandas as pd
import envelopment as env
```

```python
# Import as tabelas saneamento
# Executar somente uma fez. Depois que a tabele tenha sido escalonada import a tabela do final
df_saneamento = pd.read_csv("dados_limpos_saneamento.csv")
```

```python
inputs = np.array(df_saneamento[['Perdas Fat',
                                 'Perdas Dist',
                                 'Tarifa méd',
                                 'Investimento',
                                 'Ext rede água',
                                 'Ext rede esgotos',
                                 'Empregados']])
```

```python
outputs = np.array(df_saneamento[['At Água',
                                  'At Esgoto',
                                  'Esg Tratado',
                                  'Novas Lig Água',
                                  'Novas Lig Esgoto',
                                  'Ev Perdas Fat',
                                  'Ev Perdas Dist']])
```

```python
names = []
names += df_saneamento['Município'].tolist()
```

```python
dea = env.DEA(inputs, outputs)
```

```python
dea.name_units(names)
```

Às vezes, o algoritmo de otimização SLSQP para cedo dando o erro:

Derivativo direcional positivo para pesquisa de linhas (modo de saída 8)
Nesse caso, você pode tentar seguir as coisas:

Se você tiver os sinalizadores de compilador nativo -ffast-math -march = ativos (veja acima), tente desativá-los.
Use uma malha mais fina na área do local da turbina. Os erros numéricos de representar as turbinas podem estar dominando o problema.
Use uma tolerância de otimização mais flexível, passando o parâmetro tol para maximizar a função.
Se você estiver otimizando apenas para o atrito da turbina e não usar o parâmetro automatic_scaling (o que você deve neste caso específico), talvez seja necessário dimensionar o problema manualmente. Isto é feito passando um argumento de escala para maximizar, e. maximizar (rf, escala = 1e-3). Escolha o valor da escala de forma que o primeiro gradiente seja da ordem de 1.
